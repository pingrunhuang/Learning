```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
# plot figure without calling show() method
%matplotlib inline
# apply default seaborn theme
sns.set()
```

```python
df = pd.read_csv("data/timeseries.csv")
df.head()
```

```python
# convert the month column into datetime and set it as index
df.columns = ['month', 'diet', 'gym', 'finance']
df['month'] = pd.to_datetime(df['month'])
df.set_index('month', inplace=True)
```

```python
# plotting 3 columns in one fig
df.plot(figsize=(20, 10), linewidth=5, fontsize=20)
plt.xlabel('Year', fontsize=20)
```

```python
# plotting only one column
df[['diet']].plot(figsize=(20, 10), linewidth=5, fontsize=20)
plt.xlabel('Year', fontsize=20)
```

```python
# smoothing out
diet = df[['diet']]
# backward rolling with mean, xlabel is index by default
diet.rolling(12).mean().plot(figsize=(20, 10), linewidth=5, fontsize=20)
plt.xlabel("Year", fontsize=20)
```

```python
# smoothing out the huge drop due to seasonality
gym = df[['gym']]
# backward rolling with mean, xlabel is index by default
gym.rolling(12).mean().plot(figsize=(20, 10), linewidth=5, fontsize=20)
plt.xlabel("Year", fontsize=20)
```

```python
# mixing 2 terms in one fig
diet_gym_df = pd.concat([diet.rolling(12).mean(), gym.rolling(12).mean()], axis=1)
diet_gym_df.plot(figsize=(20, 10), linewidth=5, fontsize=20)
plt.xlabel('Year', fontsize=20)
```
