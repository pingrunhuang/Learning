```python
import pandas as pd
df = pd.DataFrame({'AAA' : [4,5,6,7], 'BBB' : [10,20,30,40],'CCC' : [100,50,-30,-50]})
df
```

```python
# if-then idiom: all the column BBB along with AAA with value larger and equal then 5 will be assigned with -1
df.loc[df.AAA >= 5,'BBB'] = -1
df
```

```python
# double column
df.loc[df.AAA >= 5,['BBB','CCC']] = 555
df
```

```python
# using a mask for filtering: notice how to gernerate a list with '*'
df_mask = pd.DataFrame({'AAA' : [True] * 4, 'BBB' : [False] * 4,'CCC' : [True,False] * 2})
df_mask
```

```python
df.where(df_mask,-1000)
df
```

```python
# if-then with numpy
import numpy as np
df = pd.DataFrame({'AAA' : [4,5,6,7], 'BBB' : [10,20,30,40],'CCC' : [100,50,-30,-50]})
df['logic'] = np.where(df['AAA'] > 5, 'high', 'low')
df
```
